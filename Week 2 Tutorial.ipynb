```julia
using Images
using ImageView
using Plots
using IJulia
using FileIO
```

```julia
rgb_image = load("self-image.jpg")
```

```julia
gray_image = Gray.(rgb_image)
```

```julia
black_white_image = (gray_image .> 0.5)
black_white_image = Gray.(black_white_image)
```

```julia
channels = channelview(float.(rgb_image))
print(size(channels))
rgb_threshold_channels = similar(channels);
```

```julia
for index in eachindex(rgb_image)
    rgb_threshold_channels[1,:,:] = (channels[1,:,:] .> 0.5)
    rgb_threshold_channels[2,:,:] = (channels[2,:,:] .> 0.5)
    rgb_threshold_channels[3,:,:] = (channels[3,:,:] .> 0.5)
end
```

```julia
rgb_threshold_image = colorview(RGB, rgb_threshold_channels)
```

```julia
p = plot(
    axis = nothing,
    layout = @layout([a b; c d]),
    size = (400,400),
    showaxis = false
)

plot!(p[1], rgb_image, ratio=1)
plot!(p[2], gray_image, ratio=1)
plot!(p[3], black_white_image, ratio=1)
plot!(p[4], rgb_threshold_image, ratio=1)
```
